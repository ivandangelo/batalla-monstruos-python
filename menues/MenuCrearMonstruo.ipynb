```python
class MenuCrearMonstruo:
    """
    
    Menu que se encarga de crear un monstruo con sus dos atributos.
    
    """
    def __init__(self, monstruo):
        self.monstruo = monstruo
        self.tipologias = {1:Tipologia.AGUA,
                           2:Tipologia.AIRE,
                           3:Tipologia.TIERRA,
                           4:Tipologia.FUEGO,}

        
    
        
    def mostrar_menu(self):
        
        
        """
        
        Guarda la tipologia principal y secundaria.No pueden ser iguales.
        
        """
        
        tipologia_principal = self.seleccionar_tipologia()
        tipologia_secundaria = self.seleccionar_tipologia()

        if tipologia_principal == tipologia_secundaria:
            print("No puede elegir las mismas tipologias.")
            self.mostrar_menu()
        else:
            self.monstruo.tipologia_principal = tipologia_principal
            self.monstruo.tipologia_secundaria = tipologia_secundaria


    def seleccionar_tipologia(self):
        
        
        """
        
        Guarda la tipologia elegida.
        
        """
        
        
        tipologia = None

        while tipologia == None:
            opcion = self.mostrar_tipologias()

            if not (opcion > 0 and opcion < 5):
                print("La opcion no es valida.")
            else:
                tipologia = self.tipologias[opcion]


        return tipologia


    def mostrar_tipologias(self):
        
        """
        Da al usuario la opcion de elegir un atributo.
        
        Excepciones:
        ValueError--Si la opcion no esta entre [1,4]
        
        """
        
        print("Ingrese el numero de la tipologia que desea utilizar: ")
        opcion = None
        while opcion == None:
            try:
                opcion = int(input("1- " + self.tipologias[1].name + "\n" + 
                              "2- " + self.tipologias[2].name + "\n" + 
                              "3- " + self.tipologias[3].name + "\n" + 
                              "4- " + self.tipologias[4].name + "\n"))
            except ValueError:
                print("La opcion no es valida.")

        return opcion
```
