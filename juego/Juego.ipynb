```python
class Juego ():
    """Se encarga de crear los personajes y mostrar el menu inicial para iniciar el juego."""
    def __init__ (self):
        self.monstruo_uno = Monstruo()
        self.monstruo_dos = Monstruo()
        self.turnoJugadorUno = True
    
    def crear_personaje(self, nombre_personaje):
        
        """Crea un personaje con nombre asociado
        
        Devuelve el personaje creado
        
        Parametros:
        nombre_personaje -- nombre que llevara el personaje a crear
        
        Excepciones:"""
        
        return Personaje(nombre_personaje)
    
    def esta_creado_el_juego(self):
        
        """
        Verifica si esta creado el juego.
        
        Devuelve el valor booleano correspondiente a si se creó el juego o no.
        Si existen 2 monstruos y 2 personajes el juego esta creado.
        
        """
        
        monstruos_creados = self.monstruo_uno and self.monstruo_dos
        personajes_creados = monstruos_creados and (self.monstruo_uno.personaje and self.monstruo_dos.personaje)
        juego_creado = monstruos_creados and personajes_creados
        
        return juego_creado
    
    def termino_el_juego(self):
        
        """
        Verifica si termino el juego.
        
        Devuelve el valor booleano correspondiente a si termino el juego o no.
        Si existen 2 monstruos y 2 personajes y alguno de los 2 monstruos tiene vida = 0
        el juego terminó.
        
        """
        
        juego_terminado = self.esta_creado_el_juego() and (self.monstruo_uno.puntos_de_vida == 0 or self.monstruo_dos.puntos_de_vida == 0)
        
        return juego_terminado
        
    def menu_ataque(self):
        while not self.termino_el_juego():
            if self.turnoJugadorUno:
                menu = MenuAtaques(self.turnoJugadorUno, self, self.monstruo_uno, self.monstruo_dos)
                menu.mostrar_menu()
            else:
                menu = MenuAtaques(self.turnoJugadorUno, self, self.monstruo_dos, self.monstruo_uno)
                menu.mostrar_menu()
                
            self.turnoJugadorUno = not self.turnoJugadorUno
            
            
    def ganador(self):
        
        """Indica cual es el ganador del juego.
        
        Devuelve el nombre del personaje ganador,si la vida de alguno de los
        dos monstruos es 0 entonces el ganador es el otro jugador.
        
        Parametros:
        
        Excepciones:
        
        """
        nombre_ganador=""
        
        if self.monstruo_uno.puntos_de_vida == 0:
            nombre_ganador = self.monstruo_dos.personaje.nombre
        elif self.monstruo_dos.puntos_de_vida == 0:
            nombre_ganador = self.monstruo_uno.personaje.nombre

        return nombre_ganador
    
    def iniciar_juego(self):
        """Inicia el juego
        Muestra el menu inicial de /menues/MenuInicial.ipynb y devuelve la opcion del menu a mostrar"""
        opcion = 0
        while opcion != 3:
            opcion = MenuInicial().mostrar_menu()
            self.elegir_menu(opcion)
            if self.termino_el_juego() and opcion != 3:
                print("Fin del juego! El ganador es: " + self.ganador())
                
        print("Se salio del juego exitosamente.")
    def elegir_menu(self, opcion):
        if self.turnoJugadorUno:
            self.hacer_menu(opcion, self.monstruo_uno, self.monstruo_dos)
        else:
            self.hacer_menu(opcion, self.monstruo_dos, self.monstruo_uno)
        self.turnoJugadorUno = not self.turnoJugadorUno
        
    def hacer_menu(self, opcion, monstruo_atacante, monstruo_atacado):
        if opcion not in [1, 2, 3]:
            print("No es una opcion valida")
        else:
            if opcion == 1:
                menu = MenuCrearJuego(self)
                menu.mostrar_menu()
            elif opcion == 2:
                menu = MenuCargar(self)
                game = menu.mostrar_menu()
                self.__dict__.update(game.__dict__)
            elif opcion == 3:
                pass
```
