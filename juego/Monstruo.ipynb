```python
class Monstruo:
    
    'Monstruo con ataque simple,4 ataques especiales, 100 puntos de vida, dos atributos.'
    
    
    ATAQUE_ESPECIAL = 15
    ATAQUE_SIMPLE = 10
    PORCENTAJE_DANIO_ATAQUE_ESPECIAL = 0.2
    
    
    def __init__(self):
        self.contador_ataques_especiales = 0
        self.puntos_de_vida = 100
        self.tipologia_principal = None
        self.tipologia_secundaria = None
        self.personaje = None
    
    
    def ataque_simple(self, monstruo_enemigo):
        
        """Ataca a otro monstruo enemigo.
        
        Usa un ataque simple y devuelve el danio aplicado
        
        Parametros:
        monstruo_enemigo -- monstruo que va a recibir el ataque.
       
        Excepciones:
        
        """  
        
        monstruo_enemigo.puntos_de_vida -= self.ATAQUE_SIMPLE

        if monstruo_enemigo.puntos_de_vida < 0:
            monstruo_enemigo.puntos_de_vida = 0
            
        return self.ATAQUE_SIMPLE
    
    
    
    
    def ataque_especial(self, tipologia_atacante, monstruo_enemigo):
        
        """Ataca con un ataque especial de "x" tipologia al monstruo enemigo.
        
        Solo se aplica el danio correspondiente si el ataque especial fue usado
        menos de 4 veces y devuelve el danio realizado.
        
        Parametros:
        tipologia_atacante -- 
        monstruo_enemigo -- monstruo que va a recibir el ataque.
        
        Excepciones:
        
        
        """
        
        

        valor_ataque = self.ATAQUE_ESPECIAL
        
        if self.contador_ataques_especiales < 4:
            if self.soy_fuerte_en_ataque(tipologia_atacante, monstruo_enemigo):
                valor_ataque += self.ATAQUE_ESPECIAL * self.PORCENTAJE_DANIO_ATAQUE_ESPECIAL
            
            if self.enemigo_es_fuerte_en_defensa(tipologia_atacante, monstruo_enemigo):
                valor_ataque -= self.ATAQUE_ESPECIAL * self.PORCENTAJE_DANIO_ATAQUE_ESPECIAL
            
            monstruo_enemigo.herir(valor_ataque)
            self.contador_ataques_especiales += 1
        else:
            valor_ataque = 0
            
        return int(valor_ataque)
    
    
    def herir(self, danio_realizado):
        
        """Aplica el danio al monstruo actual.
        
        Si el danio realizado es menor que los puntos de vida el monstruo muere.
        
        Parametros:
        danio_realizado -- danio a realizar al monstruo actual, debe ser > 10
        
        Excepciones:
        
        """
        
        
        self.puntos_de_vida -= danio_realizado

        if self.puntos_de_vida < 0:
            self.puntos_de_vida = 0
    
    
    def soy_fuerte_en_ataque(self, tipologia_atacante, monstruo_enemigo):
        
        """
        
        Indica si soy fuerte en ataque frente a la tipologia del enemigo
        
        Parametros:
        
        Excepciones:
        
        """
        
        resultado = (monstruo_enemigo.tipologia_principal.name == tipologia_atacante.value[0] or 
                    monstruo_enemigo.tipologia_secundaria.name == tipologia_atacante.value[0])
        
        
        return resultado    
    
    def enemigo_es_fuerte_en_defensa(self, tipologia_atacante, monstruo_enemigo):
        
        """
        
        Indica si el enemigo es fuerte frente a mi tipologia
        
        Parametros:
        
        Excepciones:
        
        """
        
        resultado = (monstruo_enemigo.tipologia_principal.value[1] == tipologia_atacante.name or 
                    monstruo_enemigo.tipologia_secundaria.value[1] == tipologia_atacante.name)
        
        return resultado
```
